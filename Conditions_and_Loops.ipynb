```python
""" 1.  What are conditional statements? Explain conditional statements with syntax and examples.

    Conditional statements in JavaScript allow you to execute certain blocks of code only when specific conditions are met. They are fundamental for decision-making in your programs.

     1. if Statement: Executes code only if the condition is true.

          Syntax:
          if (condition) {
            // code to execute if condition is true
          }

          Example:
          let age = 20;
          if (age >= 18) {
            console.log("You are eligible to vote.");
          }

    2. if else Statement: Executes one block if the condition is true, and another block if it's false.

          Syntax:
          if (condition) {
            // code if condition is true
          } else {
            // code if condition is false
          }

          Example:
          let marks = 40;
          if (marks >= 50) {
            console.log("Pass");
          } else {
            console.log("Fail");
          }

    3. if else if else Ladder: Checks multiple conditions in sequence.

          Syntax:
          if (condition1) {
            // if condition1 is true
          } else if (condition2) {
            // if condition2 is true
          } else {
            // if none of the above are true
          }

          Example:
          let num = 0;
          if (num > 0) {
            console.log("Positive");
          } else if (num < 0) {
            console.log("Negative");
          } else {
            console.log("Zero");
          }

    4. Nested if Statement: An if statement placed inside another if.

          Example:
          let age = 25;

          if (age > 18) {
            if (age < 60) {
              console.log("You are an adult.");
            }
          }

     5. switch Statement: Used to check multiple values of a variable more cleanly.

          Syntax:
          switch(expression) {
            case value1:
              // code block
              break;
            case value2:
              // code block
              break;
            default:
              // default code block
          }

          Example:
          let day = 3;

          switch (day) {
            case 1:
              console.log("Monday");
              break;
            case 2:
              console.log("Tuesday");
              break;
            case 3:
              console.log("Wednesday");
              break;
            default:
              console.log("Invalid day");
          }
```

```python
""" 2. Write a program that grades students based on their marks
        1. If greater than 90 then A Grad
        2. If between 70 and 90 then a B grad
        3. If between 50 and 70 then a C grad
        4. Below 50 then an F grade """


      let marks = 85;

      if (marks > 90) {
        console.log("Grade: A");
      } else if (marks >= 70 && marks <= 90) {
        console.log("Grade: B");
      } else if (marks >= 50 && marks < 70) {
        console.log("Grade: C");
      } else {
        console.log("Grade: F");
      }
```

```python
""" 3.  What are loops, and what do we need them? Explain different types of loops with their syntax and
examples.

    Loops are control structures that allow a block of code to repeat multiple times based on a condition.
    They are essential when you need to perform repetitive tasks like iterating over arrays, counting, or performing repeated calculations.

    We Need Loops
      To avoid writing repetitive code
      To process collections (like arrays or lists)
      To automate repetitive tasks (e.g., printing 1 to 100)

    Types of Loops:
    1. for Loop: Used when you know in advance how many times to run the loop.
        Syntax:
        for (initialization; condition; increment) {
          // code block to be executed
        }

        Example:
        for (let i = 1; i <= 5; i++) {
          console.log("Number:", i);
        }

    2. while Loop: Used when the number of iterations is unknown — runs as long as the condition is true.
        Syntax:
        while (condition) {
          // code block
        }

        Example:
        let i = 1;
        while (i <= 3) {
          console.log("Count:", i);
          i++;
        }

    3. do...while Loop: Like while, but guarantees the loop runs at least once, even if the condition is false at the start.
        Syntax:
        do {
          // code block
        } while (condition);

        Example:
        let i = 1;
        do {
          console.log("Value:", i);
          i++;
        } while (i <= 2);

    4. for...of Loop (JavaScript specific): Used to iterate over iterable objects like arrays.
        Example:
        let fruits = ["Apple", "Banana", "Cherry"];
        for (let fruit of fruits) {
          console.log(fruit);
        }
```
